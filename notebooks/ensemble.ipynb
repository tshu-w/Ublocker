```python
from sklearn.metrics import auc

result_dir = Path("./results/debug/")


def evaluate(
    candidates: list[set[tuple[Any, Any]]],
    matches: set[tuple[Any, Any]],
    *,
    threshold: float = 0.9,
) -> dict:
    precisions, recalls = [1], [0]
    cands = set()

    average_precision = 0
    for i in range(len(candidates)):
        cands = cands | candidates[i]

        tp = len(cands & matches)
        precision = tp / len(cands)
        recall = tp / len(matches)
        average_precision += precision * (recall - recalls[-1])

        precisions.append(precision)
        recalls.append(recall)

    k = 0
    for i in range(len(candidates)):
        precision = precisions[i]
        recall = recalls[i]
        k = i
        if recall > threshold:
            break

    average_precision_2 = auc(recalls, precisions)
    
    return {
        "AP": average_precision * 100,
        "AP2": average_precision_2 * 100,
        "PC": recall * 100,
        "PQ": precision * 100,
        "F1": 2 * (precision * recall) / (precision + recall) * 100,
        "K": float(k),
    }, recalls, precisions

for d in data_dirs[1:]:
    if d.name != "walmart-amazon_heter":
        continue
    print(d.name)
    matches_path = Path(d) / "matches.csv"
    matches = set(
        pd.read_csv(matches_path).itertuples(index=False, name=None)
    )
    with (result_dir / "sparse_join" / f"{d.name}.pickle").open("rb") as f:
        candidates_sparse = pickle.load(f)

    with (result_dir / "dense" / f"{d.name}.pickle").open("rb") as f:
        candidates_dense = pickle.load(f)

    flag = set()
    candidates = []
    for s1, s2 in zip(candidates_sparse, candidates_dense):
        s = s1 | s2
        s = s - flag
        candidates.append(s)
        flag |= s

    prs = []
    metrics, recalls, precisions = evaluate(candidates, matches)
    print(" & ".join(map(str, metrics.values())))
    for i, (r, p) in enumerate(zip(recalls, precisions)):
        prs.append({
            "Data": d.name,
            "Type": "ensemble",
            "Recall": r,
            "Precision": p,
            "k": i,
        })
        
    metrics, recalls, precisions = evaluate(candidates_dense, matches)
    print(" & ".join(map(str, metrics.values())))
    for i, (r, p) in enumerate(zip(recalls, precisions)):
        prs.append({
            "Data": d.name,
            "Type": "dense",
            "Recall": r,
            "Precision": p,
            "k": i,
        })
    
    metrics, recalls, precisions = evaluate(candidates_sparse, matches)
    print(" & ".join(map(str, metrics.values())))
    for i, (r, p) in enumerate(zip(recalls, precisions)):
        prs.append({
            "Data": d.name,
            "Type": "sparse",
            "Recall": r,
            "Precision": p,
            "k": i,
        })
    
    df = pd.DataFrame(prs)
    sns.relplot(data=df, x="Recall", y="Precision", kind="scatter", hue="Type").set(title=f"{d.name}")
```
